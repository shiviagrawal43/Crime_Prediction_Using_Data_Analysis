```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn import linear_model
import pickle
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('intial_data_set.csv')
df.head()
```

```python
train, test = train_test_split(df, test_size=0.5)
```

```python
xtest = pd.DataFrame(test, columns=['Year','Month','City','Crime'])
ytest = pd.DataFrame(test, columns=['Frequency'])
xtrain= pd.DataFrame(train, columns=['Year','Month','City','Crime'])
ytrain = pd.DataFrame(train, columns=['Frequency'])
```

```python
xtest = pd.get_dummies(data=xtest, drop_first=True)
xtrain = pd.get_dummies(data=xtrain, drop_first=True)
```

```python
model = sm.OLS(ytrain.astype(float),xtrain.astype(float)).fit()
predictions = model.predict(xtest.astype(float))
model.summary()
```
